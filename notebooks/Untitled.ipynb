```python
sort_dt= data.sort_values(by=['Patient ID', 'Lab Resulted Order Test Description', 'Specimen Received Date'])
sort_dt
```

```python
sort_dt['test_number'] = sort_dt.groupby('Patient ID').cumcount() + 1

final_concat = sort_dt.pivot(index= 'Patient ID', columns = 'test_number', values = ['Result Value', 'Specimen Received Date', 'Lab Resulted Order Test Description'])

final_concat.columns = [f"{test}_{i}" for test, i in final_concat.columns]
final_concat = final_concat.reset_index()
final_concat
```
